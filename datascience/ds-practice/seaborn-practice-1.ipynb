```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.info()
tips.head()
```

```python
sns.distplot(tips["total_bill"],bins=30)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg")
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.kdeplot(tips["total_bill"])
```

```python
import numpy as np
```

# Categorial plots

```python
sns.barplot(x='day',y="tip",data=tips,hue='sex')
```

```python
sns.barplot(x='day',y="tip",data=tips, estimator=np.std,hue='sex')
```

```python
tips.describe()
```

```python
sns.countplot(x='day', data=tips)
```

```python
sns.countplot(x='day',data=tips,hue='sex')
```

```python
sns.boxplot(x='day',y='tip',data=tips)
```

```python
sns.boxplot(x='day',y='tip',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='tip',data=tips,hue='sex', split=True)
```

```python
sns.stripplot(x='day',y='tip',data=tips, jitter=True,hue='sex',dodge=True)
```

```python

sns.swarmplot(x='day',y='tip',data=tips,hue='sex', dodge=True)
```

```python
sns.violinplot(x='day',y='tip',data=tips,hue='sex' , split=False)
sns.swarmplot(x='day',y='tip',data=tips,hue='sex', split=True)
```

```python
sns.catplot(x='day',y='tip',data=tips,hue='sex',kind='bar',estimator=np.std)
```
